```python
import os

from dotenv import load_dotenv

load_dotenv(os.path.join("..", ".env"), override=True)
```

```python
import os
import sys
import subprocess
import time

# 1. Absolute path to your src directory
src_path = os.path.abspath("src")
sys.path.append(src_path)

# 2. Prepare environment variables
env = os.environ.copy()
env["PYTHONPATH"] = src_path + os.pathsep + env.get("PYTHONPATH", "")

# 3. Start NAT without the --tools flag
nat_process = subprocess.Popen(
    [
        "nat", "serve", 
        "--config_file", "./configs/config.yml", 
        "--host", "127.0.0.1", 
        "--port", "8000"
    ],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True,
    env=env
)

print("â³ Waiting for NAT server to initialize...")
time.sleep(12)

# Check if it crashed
if nat_process.poll() is not None:
    stdout, stderr = nat_process.communicate()
    print("âŒ Server crashed!")
    print("\n=== Error Output ===")
    print(stderr)
else:
    print("âœ… Server is running!")
```

```python
from pathlib import Path

ui_path = Path("NeMo-Agent-Toolkit-UI")

if not ui_path.exists():
    print("Cloning NAT UI repository...")
    subprocess.run([
        "git", "clone",
        "https://github.com/NVIDIA/NeMo-Agent-Toolkit-UI.git"
    ], check=True)
    print("UI repository cloned")
else:
    print("UI respository already exists!")
```

```python
# uncomment and run if you are running in your own environment
print("Installing dependencies...this could take a few minutes")
subprocess.run(["npm", "ci"], cwd=ui_path, check=True)
print("Dependencies installed")
```

```python
# start UI
print("Starting UI development server...")
ui_process = subprocess.Popen(
    ["npm", "run", "dev"],
    cwd=ui_path,
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True,
    env={**os.environ, "NEXT_TELEMETRY_DISABLED": "1"}
)

time.sleep(30)
print(f"UI started!")
```

```python
print("Open the NAT UI at: http://localhost:3000")
print("NAT backend running at: http://localhost:8000")
```
